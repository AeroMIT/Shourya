```python
#finding out the hsv values of the purple and red in the image using the bar of hsv values


import cv2
import numpy as np

def nothing(x):
    pass

# Load image
image = cv2.imread(x)
image = cv2.resize(image, (400,400))
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


cv2.namedWindow('HSV Adjustments')

cv2.createTrackbar('H Low', 'HSV Adjustments', 0, 179, nothing)
cv2.createTrackbar('H High', 'HSV Adjustments', 179, 179, nothing)
cv2.createTrackbar('S Low', 'HSV Adjustments', 0, 255, nothing)
cv2.createTrackbar('S High', 'HSV Adjustments', 255, 255, nothing)
cv2.createTrackbar('V Low', 'HSV Adjustments', 0, 255, nothing)
cv2.createTrackbar('V High', 'HSV Adjustments', 255, 255, nothing)

while True:
    
    h_low = cv2.getTrackbarPos('H Low', 'HSV Adjustments')
    h_high = cv2.getTrackbarPos('H High', 'HSV Adjustments')
    s_low = cv2.getTrackbarPos('S Low', 'HSV Adjustments')
    s_high = cv2.getTrackbarPos('S High', 'HSV Adjustments')
    v_low = cv2.getTrackbarPos('V Low', 'HSV Adjustments')
    v_high = cv2.getTrackbarPos('V High', 'HSV Adjustments')
    
    
    lower_bound = np.array([h_low, s_low, v_low])
    upper_bound = np.array([h_high, s_high, v_high])
    mask = cv2.inRange(image_hsv, lower_bound, upper_bound)
    
    
    result = cv2.bitwise_and(image, image, mask=mask)
    
   
    cv2.imshow('Original', image)
    cv2.imshow('Mask', mask)
    cv2.imshow('Filtered Image', result)
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
